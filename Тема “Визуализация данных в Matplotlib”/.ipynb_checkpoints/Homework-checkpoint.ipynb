```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

Task 1

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
title_dict = {'fontsize': 18, 'fontweight': 'bold', 'family': 'Calibri'}
label_dict = {'fontsize': 14, 'color': 'Red', 'family': 'Calibri'}
```

```python
plt.plot(x, y)

plt.title('График', fontdict=title_dict)

plt.xlabel('x', fontdict=label_dict)
plt.ylabel('y', fontdict=label_dict)

plt.show()
```

```python
plt.scatter(x, y)

plt.title('График', fontdict=title_dict)

plt.xlabel('x', fontdict=label_dict)
plt.ylabel('y', fontdict=label_dict)

plt.grid(color='ghostwhite')

plt.show()
```

Task 2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
plt.plot(t, f, color='Green')

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.title('График f(t)', fontdict=title_dict)

plt.xlabel('Значения t', fontdict=label_dict)
plt.ylabel('Значения f', fontdict=label_dict)

plt.grid(color='grey')

plt.show()
```

Task 3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1', fontdict=title_dict)
ax1.axis([-5, 5, 0, 10])

ax2.plot(x, y2)
ax2.set_title('График y2', fontdict=title_dict)

ax3.plot(x, y3)
ax3.set_title('График y3', fontdict=title_dict)

ax4.plot(x, y4)
ax4.set_title('График y4', fontdict=title_dict)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
legend_dict = {'family': 'Calibri', 'size': 10}
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
data_class = data['Class']
data_class.value_counts()
```

```python
hist_class = plt.hist(data_class, bins=2)
plt.show()
```

```python
plt.hist(data.loc[data['Class'] == 1]['V1'], bins=20, alpha=0.5, color='Red', density=True, label='Class 1')
plt.hist(data.loc[data['Class'] == 0]['V1'], bins=20, alpha=0.5, color='Grey', density=True, label='Class 0')

plt.xlabel('Class', fontdict=label_dict)
plt.yscale(value='log')
plt.legend(prop=legend_dict)
plt.show()
```

```python
#12
C = g11.reshape((4, 4))
D = np.add((C*10), B)
det_D = np.linalg.det(D)
D_inv = np.linalg.inv(D)
D_rank = np.linalg.matrix_rank(D)
print(det_D, D_rank)
print(D_inv)
```

```python
#13
for indexes, x in np.ndenumerate(D_inv):
    if x < 0:
        D_inv[indexes] = 0
    else:
        D_inv[indexes] = 1
D_inv
        
```
